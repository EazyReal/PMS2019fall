```python
import numpy as np
```

```python
A = np.eye(12)*4
for i in range(0,3):
    A[i][i+1] = A[i+1][i] = -1
for i in range(0,3):
    A[11-i][10-i] = A[10-i][11-i] = -1
for i in range(0,4):
    A[5+i][3+i] = A[3+i][5+i] = -1 
A[4][0] = A[0][4] = -1
A[7][11] = A[11][7] = -1
print(A)
b = np.asarray([220, 110, 110, 220, 110, 110, 110, 110, 220, 110, 110, 220]).T

eps = 1e-5
x0 = np.zeros(A.shape[0])
```

```python
Ainv = np.linalg.inv(A)
print("direct solution:")
print(np.dot(Ainv, b))
```

```python
def Jacobi(A, b, x0=x0, eps=eps, N=30):
    print("Solving Ax = b with Jacobi method...")                                                                                                                                                      
    if x0 is None:
        x0 = np.zeros(A.shape[0])
                                                                                                                                                                
    M = np.diagflat(np.diag(A))
    AM = A - M                                                                                                                                                              
    for i in range(N):
        if(np.linalg.norm((np.dot(A,x0)-b)) < eps):
            print("norm Ax-b < eps @ iteration {}".format(i))
            return x0
        print("interation {}:\n".format(i), x0)
        x1 = np.dot(np.linalg.inv(M), b - np.dot(AM, x0)) #in here
        x0 = x1
    return x0
```

```python
Jacobi(A,b)
```

```python
def Gauss_Seidel(A, b, x0=x0, eps=eps, N=30):
    print("Solving Ax = b with Gauss Seidel method...")                                                                                                                                                      
    if x0 is None:
        x0 = np.zeros(A.shape[0])

    M = np.tril(A)
    AM = A - M
    for i in range(N):
        if(np.linalg.norm((np.dot(A,x0)-b)) < eps):
            print("norm Ax-b < eps @ iteration {}".format(i))
            return x0
        print("interation {}:\n".format(i), x0)
        x1 = np.dot(np.linalg.inv(M), b - np.dot(AM, x0))
        x0 = x1
    return x0
```

```python
Gauss_Seidel(A,b)
```

```python
def SOR(A, b, w, x0=x0, eps=eps, N=30, show=True):
    if show:
        print("Solving Ax = b with SOR method, w = {}...".format(w))                                                                                                                                                      
    if x0 is None:
        x0 = np.zeros(A.shape[0])

    D = np.diagflat(np.diag(A))
    L = np.tril(A)-D
    M = D+w*L
    AM = A - M                   
    for i in range(N):
        if(np.linalg.norm((np.dot(A,x0)-b)) < eps):
            if show:
                print("norm Ax-b < eps @ iteration {}".format(i))
            return x0, i
        if show:
            print("interation {}:\n".format(i), x0)
        x1 = np.dot(np.linalg.inv(M), b - np.dot(AM, x0))
        x0 = x1
    return x0, N
```

```python
SOR(A, b, 1.0862172434486896) #1.1-1.3, 9it, 1.1 13 1e-5
```

```python
#run to find minw for SOR
ws = np.linspace(0.0, 2.0, 5000)
minit = 30;
minw = 0.0;
for wi in ws:
    xi, ri = SOR(A, b, wi, show=False)
    if(ri < minit):
        minw = wi
        minit = ri
    
```
